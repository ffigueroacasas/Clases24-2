```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los datos
clientes = pd.read_csv("clientes.csv")
destinatarios = pd.read_csv("Destinatarios.txt", sep=",", header=None, encoding="latin-1")  # ajustar separador si es necesario
destinatarios.columns = ["IdCiudad","Nombre","Apellido","FechaNacimiento","EstadoCivil","Genero","Email","IngresoAnual","TotalHijos","CantHijosEnCasa","Educacion","Ocupacion","Propietario","CantAutomoviles","Direccion","Telefono","Distancia","Region","Edad"]

#La primer fina de destinatarios contiene los nombres de columnas
destinatarios.drop(0, axis=0, inplace=True)
```

```python
# Exploración inicial
print(clientes.info())
print(clientes.describe())
```

```python
# Verificar valores faltantes
print(clientes.isnull().sum())
```

```python
# Análisis univariante
# Para variables numéricas
for col in clientes.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(10, 4))
    plt.subplot(1, 2, 1)
    sns.histplot(clientes[col], kde=True)
    plt.title(f'Distribución de {col}')

    plt.subplot(1, 2, 2)
    sns.boxplot(y=clientes[col])
    plt.title(f'Boxplot de {col}')
    plt.tight_layout()
    plt.show()

# Para variables categóricas
for col in clientes.select_dtypes(include=['object']).columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(y=clientes[col])
    plt.title(f'Frecuencia de {col}')
    plt.tight_layout()
    plt.show()
```

```python
# Análisis bivariante con la variable objetivo
target = 'ComproBicicleta'
# Para variables numéricas vs target
for col in clientes.select_dtypes(include=['int64', 'float64']).columns:
    if col != target:
        plt.figure(figsize=(10, 5))
        sns.boxplot(x=clientes[target], y=clientes[col])
        plt.title(f'{col} vs {target}')
        plt.show()
```

```python
# Para variables categóricas vs target
for col in clientes.select_dtypes(include=['object']).columns:
    if col != target:
        plt.figure(figsize=(10, 6))
        pd.crosstab(clientes[col], clientes[target]).plot(kind='bar')
        plt.title(f'{col} vs {target}')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
```
